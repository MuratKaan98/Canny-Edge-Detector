```python
%matplotlib inline


import utils #for directing the 'test_imgs' folder
import CEDetector as ced #Necessary steps to perform  the canny edge detection algorithm

```

```python
imgs = utils.load_data()
utils.visualize(imgs, 'gray')

#importing data from 'test_imgs' folder
```

```python
detector = ced.cannyEdgeDetector(imgs, sigma=1.4, kernel_size=5, lowthreshold=0.09, highthreshold=0.17, weak_pixel=100)

#Sigma = 1.4
#Kernel Size = 5x5

#detector function actually performs these actions step by step as defined in ced.cannyEdgeDetector

#smoothing = convolving operation
#calculating gradient            
#Non-max calculations
#Thresholding
           
```

```python
imgs_final = detector.detect()
#Does all of the operations above
```

```python
utils.visualize(imgs_final, 'gray') #Visualize
```
